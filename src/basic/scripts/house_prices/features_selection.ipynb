```python
import pandas
data = pandas.read_csv('data/train.csv')
data.corr()
```

```python
correlation = data.corr()
sale_price_correlation = correlation.SalePrice
minimum_covariance = 0.1
good_columns = sale_price_correlation[lambda x: abs(x) >= minimum_covariance].index.values
good_columns = [c for c in good_columns]
```

```python
data = data[good_columns]
```

```python
data.head()
```

```python
X = data.select_dtypes(exclude=['object']).drop('SalePrice', axis=1)
y = data.SalePrice
X.columns
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer(strategy='mean')
X_imputed = my_imputer.fit_transform(X)
X_imputed.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
        X_imputed, y, test_size=0.2)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=10000)
model = model.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn import metrics
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy
from sklearn.impute import SimpleImputer

data = pandas.read_csv('data/train.csv')
X = data.select_dtypes(exclude=['object']).drop('SalePrice', axis=1)
y = data.SalePrice
my_imputer = SimpleImputer(strategy='mean')
X_imputed = my_imputer.fit_transform(X)
X_categorical = data.select_dtypes(include=['object'])
my_imputer = SimpleImputer(strategy='most_frequent')
X_categorical_imputed = my_imputer.fit_transform(X_categorical)
my_imputer = SimpleImputer(strategy='most_frequent')
X_categorical_imputed = my_imputer.fit_transform(X_categorical)
encoder = OneHotEncoder(handle_unknown='ignore')
encoded = encoder.fit(X_categorical_imputed)
transformed = encoded.fit_transform(X_categorical_imputed)
merged = numpy.concatenate((X_imputed, transformed.toarray()), axis=1)
merged_columns = [c for c in X.columns] + [c for c in encoder.get_feature_names()]
len(merged_columns)
```

```python
# Variance Threshold
```

```python
merged_data_frame = pandas.DataFrame(data=merged, columns=merged_columns)
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.05 * (1 - .05)))
```

```python
from sklearn import preprocessing

selected = selection.fit_transform(merged_data_frame)
scaler = preprocessing.StandardScaler()

selected.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(
        selected, y, test_size=0.2)
x_train, x_test = scaler.fit_transform(x_train), scaler.fit_transform(x_test)
model = XGBRegressor(n_estimators=10000)
model = model.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_predict = model.predict(x_test)
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```

```python
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression
selection = SelectKBest(f_regression, k=100)
selected = selection.fit_transform(merged_data_frame, y)
x_train, x_test, y_train, y_test = train_test_split(
        selected, y, test_size=0.2)
x_train, x_test = scaler.fit_transform(x_train), scaler.fit_transform(x_test)
model = XGBRegressor(n_estimators=10000)
model = model.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_predict = model.predict(x_test)
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```

```python
selection = SelectKBest(mutual_info_regression, k=100)
selected = selection.fit_transform(merged_data_frame, y)
x_train, x_test, y_train, y_test = train_test_split(
        selected, y, test_size=0.2)
x_train, x_test = scaler.fit_transform(x_train), scaler.fit_transform(x_test)
model = XGBRegressor(n_estimators=10000)
model = model.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_predict = model.predict(x_test)
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```
