```python
import pandas 
```

```python
data = pandas.read_csv('data/train.csv')
data.head()
```

```python
data.columns
```

```python
# X will contain only the numeric features of the dataset and excludes the target variable
```

```python
X = data.select_dtypes(exclude=['object']).drop('SalePrice', axis=1)
y = data.SalePrice
X.columns
```

```python
# We replace all missing values with the mean value of the existing values
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer(strategy='mean')
X_imputed = my_imputer.fit_transform(X)
X_imputed.shape
```

```python
# We fill the values of the categorical variables with the most frequent value
```

```python
X_categorical = data.select_dtypes(include=['object'])
my_imputer = SimpleImputer(strategy='most_frequent')
X_categorical_imputed = my_imputer.fit_transform(X_categorical)
X_categorical.columns
```

```python

my_imputer = SimpleImputer(strategy='most_frequent')
X_categorical_imputed = my_imputer.fit_transform(X_categorical)
X_categorical_imputed.shape
```

```python
# Encoding the categorical values
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy
encoder = OneHotEncoder(handle_unknown='ignore')
encoded = encoder.fit(X_categorical_imputed)
transformed = encoded.fit_transform(X_categorical_imputed)
transformed.shape
```

```python
merged = numpy.concatenate((X_imputed, transformed.toarray()), axis=1)
merged.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
        merged, y, test_size=0.2)
```

```python
model = RandomForestRegressor(random_state=1, n_estimators=x_train.shape[1])
```

```python
model = model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn import metrics
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=10000)
model = model.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_predict = model.predict(x_test)
```

```python
score_sklearn = metrics.mean_absolute_error(y_predict, y_test)
score_sklearn
```
